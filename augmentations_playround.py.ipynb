```python
import albumentations as A
```

```python
import cv2
```

```python
transform = A.Compose([
    A.GaussNoise(),
    A.HorizontalFlip(p=0.5, always_apply=True),
    A.RandomBrightnessContrast(p=0.2, always_apply=True),
])
```

```python
image = cv2.imread("/Users/pascal/Downloads/hparam_k.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
from matplotlib import pyplot as plt


def visualize(image):
    plt.figure(figsize=(10, 10))
    plt.axis('off')
    plt.imshow(image)
```

```python
for t in transform.transforms:
    transformed_image = t(image=image)["image"]
    print(str(type(t)).split("'")[1].split(".")[-1])
    visualize(transformed_image)
    plt.show()
```
